**Short Rate Models**

\begin{align*}
\text{Vasicek:} &\qquad dr_t = (b-ra_t)dt + \sigma dW_t \\
\text{CIR:} &\qquad dr_t = a(b-r_t)dt + \sigma \sqrt{r_t} dW_t \\
\text{Ho-Lee:} &\qquad dr_t = \theta(t)dt + \sigma dW_t \\
\text{Hull-White:} &\qquad dr_t = (\theta(t)-ar_t)dt + \sigma \sqrt{r_t} dW_t
\end{align*}

Vasicek:
- Mean reverting
- Allows for negative rates
- Gaussian distribution
- Closed form solution
- Bound settels at a certian level
- Explicit formula for the ZCB
- Pricing option prices on ZCBs and par swap rates

Suppose we have a cap of caplets with $\text{payoff: }(L(T_{i-1},T_i)-K)_+$ where $L$ is the caplet. We can just rewrite the payoff to be a replicating model of an option portfolio with the ZCB as the underlying $(P(T_{i-1},T_i)-K)_+$. 

Cox-Ingersoll-Ross:
- Cannot be solved in closed form
- You could find the european option price in closed form

Ho-Lee / Hull-White Extended Vasicek:
- You 'input' the the initial term structure such that it fits perfectly in $\theta(t)$. Where CIR and Vasicek have a limited number of parameters (only alow to fit strictly increasing or decreasing term structures).
- We can get prices on ZCBs and caplet options
